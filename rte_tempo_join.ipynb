```python
import pandas as pd
```

```python
def rte_tempo_join(data_rte_df, tempo_df, drop_last=False):
    if drop_last:
        tempo_df.drop(tempo_df.tail(1).index,inplace=True)

    #Format date
    data_rte_df["Date"] = pd.to_datetime(data_rte_df['Date'], format='%Y-%m-%d')
    tempo_df['Date'] =  pd.to_datetime(tempo_df['Date'], format='%Y-%m-%d %H:%M:%S')
    return pd.merge(data_rte_df, tempo_df, on="Date", how="left")
```

```python
data_rte = pd.read_csv("clean_datasets/rte_production.csv").reset_index()
```

```python
tempo = pd.read_csv("clean_datasets/tempo_2014_2024.csv")
```

```python
merge = rte_tempo_join(data_rte, tempo)
merge.to_csv("clean_datasets/rte_2014_2024.csv")
```
